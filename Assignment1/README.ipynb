# Assignment 1 — SMS 分类 (简要说明)

简介  
本仓库的 Assignment1 工作在 [Assignment1/Assignment1-Doc.ipynb](Assignment1/Assignment1-Doc.ipynb)。任务是从短信文本预测三类标签：`0=normal`, `1=spam`, `2=smishing`（短信钓鱼）。

快速开始
- 打开并运行笔记本：[Assignment1/Assignment1-Doc.ipynb](Assignment1/Assignment1-Doc.ipynb)。
- 相关数据文件：
  - 训练数据: [Assignment1/smishing_train.txt](Assignment1/smishing_train.txt)
  - 验证/测试文本: [Assignment1/smishing_val_test.txt](Assignment1/smishing_val_test.txt)
  - 验证/测试标签: [Assignment1/smishing_val_test.csv](Assignment1/smishing_val_test.csv)
  - 示例提交: [Assignment1/my_submission.csv](Assignment1/my_submission.csv), [Assignment1/my_submission_final.csv](Assignment1/my_submission_final.csv)

主要步骤（笔记本内）
- 加载与示例工具：使用 [`read_text_data`](Assignment1/Assignment1-Doc.ipynb) 读取文本，使用 [`write_csv`](Assignment1/Assignment1-Doc.ipynb) 生成提交文件。
- 文本规范化与衍生特征：核心函数 [`normalize_text`](Assignment1/Assignment1-Doc.ipynb) 和 [`make_numeric_features`](Assignment1/Assignment1-Doc.ipynb)（长度、数字比例、URL/电话号码特征）。
- 向量化：词级/字符级 TF‑IDF（变量 [`word_vect`](Assignment1/Assignment1-Doc.ipynb), [`char_vect`](Assignment1/Assignment1-Doc.ipynb)），产生 [`trainX_tfidf`](Assignment1/Assignment1-Doc.ipynb), [`trainX`](Assignment1/Assignment1-Doc.ipynb), [`valX`](Assignment1/Assignment1-Doc.ipynb), [`testX`](Assignment1/Assignment1-Doc.ipynb)。
- 超参搜索与评估：使用 [`run_grid_search`](Assignment1/Assignment1-Doc.ipynb) 进行交叉验证，评估指标为 `balanced_accuracy_score`（见笔记本内的 [`balanced_accuracy_score`](Assignment1/Assignment1-Doc.ipynb)）。
- 试验过的模型与变量（可在笔记本中定位）：
  - MultinomialNB: [`mnb_gs`](Assignment1/Assignment1-Doc.ipynb), [`mnb_best`](Assignment1/Assignment1-Doc.ipynb)
  - LogisticRegression: [`lr_gs`](Assignment1/Assignment1-Doc.ipynb), [`lr_best`](Assignment1/Assignment1-Doc.ipynb)
  - LinearSVC: [`lsvc_gs`](Assignment1/Assignment1-Doc.ipynb), [`lsvc_best`](Assignment1/Assignment1-Doc.ipynb)
  - SGDClassifier: [`sgd_gs`](Assignment1/Assignment1-Doc.ipynb), [`sgd_best`](Assignment1/Assignment1-Doc.ipynb)

生成提交
- 选定在验证集上表现最好的模型后，在测试子集上调用模型预测并用 [`write_csv`](Assignment1/Assignment1-Doc.ipynb) 写出 CSV（格式为 Id,Prediction），示例输出为 [Assignment1/my_submission.csv](Assignment1/my_submission.csv)。

提示与注意事项
- 避免在预处理或向量化时泄露测试集信息：仅在训练集上 fit 向量化器/缩放器，然后对验证/测试做 transform（笔记本已实现）。
- 对类别不平衡可使用 class_weight 或重采样策略；评估以平衡准确率为准。
- 若使用深度学习（笔记本展示了 TextCNN 示例），请在笔记本中查看 Tokenizer / padding 的拟合位置（只在训练集上拟合）。

更多细节与实现请阅读笔记本：[Assignment1/Assignment1-Doc.ipynb](Assignment1/Assignment1-Doc.ipynb)（包含代码、超参搜索记录与结果汇总）。
