# Tutorial 3 — Predicting Bike Sharing Usage

简要说明  
本教程在 [Tutorial3/Tutorial3.ipynb](Tutorial3/Tutorial3.ipynb) 中实现了用于预测每日共享单车使用量的端到端流程：数据加载、特征工程（独热编码）、基线对照、线性/正则化回归、特征选择、以及多种非线性回归和集成方法（包括 XGBoost）。

快速开始
- 在包含 `Bike-Sharing-Dataset/day.csv` 的目录中打开并运行 [Tutorial3/Tutorial3.ipynb](Tutorial3/Tutorial3.ipynb)。
- 主要依赖：pandas、numpy、scikit-learn、matplotlib、xgboost、scipy。
- 运行顺序：加载数据 -> 特征预处理（独热编码） -> 划分按年训练/测试 -> 基准 Dummy -> 线性/岭回归 -> LASSO/OMP 特征选择 -> 非线性回归与模型选择 -> 处理测试集偏差。
