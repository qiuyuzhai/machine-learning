# Tutorial 2 — Gender Classification from Face Images

简要说明  
本教程在 [Tutorial2/Tutorial2.ipynb](Tutorial2/Tutorial2.ipynb) 中实现了基于人脸图像的性别分类实验（逻辑回归、SVM、裁剪人脸、以及自定义核 SVM）。

主要文件  
- Notebook: [Tutorial2/Tutorial2.ipynb](Tutorial2/Tutorial2.ipynb)

快速说明与运行步骤
- 在包含 `photos-bw.zip` 的目录打开并运行 [Tutorial2/Tutorial2.ipynb](Tutorial2/Tutorial2.ipynb)。
- 主要依赖：scikit-learn、scipy、numpy、matplotlib。
- 执行顺序：加载数据 -> 展示 & 展平图像 -> 预处理 (X/Y, avgX) -> 划分数据 (trainX/testX/trainY/testY) -> 训练 LR 与 SVM -> 可视化权重与误分类 -> 裁剪人脸后重复实验 -> 自定义核 SVM。

关键变量 / 函数（在 notebook 中）
- 数据与预处理：[`imgdata`](Tutorial2/Tutorial2.ipynb), [`genders`](Tutorial2/Tutorial2.ipynb), [`X`](Tutorial2/Tutorial2.ipynb), [`Y`](Tutorial2/Tutorial2.ipynb), [`imgsize`](Tutorial2/Tutorial2.ipynb), [`avgX`](Tutorial2/Tutorial2.ipynb)
- 显示与拼接：[`image_montage`](Tutorial2/Tutorial2.ipynb)
- 划分与训练：[`trainX`](Tutorial2/Tutorial2.ipynb), [`testX`](Tutorial2/Tutorial2.ipynb), [`trainY`](Tutorial2/Tutorial2.ipynb), [`testY`](Tutorial2/Tutorial2.ipynb)
- 裁剪相关：[`imgmask`](Tutorial2/Tutorial2.ipynb), [`Xm`](Tutorial2/Tutorial2.ipynb), [`masksize`](Tutorial2/Tutorial2.ipynb)
- 模型与自定义核：[`logreg`](Tutorial2/Tutorial2.ipynb), [`svm`](Tutorial2/Tutorial2.ipynb), [`best_model`](Tutorial2/Tutorial2.ipynb), [`kernel_func`](Tutorial2/Tutorial2.ipynb)

自定义核（笔记中使用的 L1-based RBF-like 核）  
$$
k(\mathbf{x},\mathbf{y}) = \exp\left(-\alpha \sum_{i=1}^d |x_i - y_i|\right)
$$
