# Tutorial 4 — Face Detection with MLPs

简要说明  
本教程在 [Tutorial4/Tutorial4 _后加了直方.ipynb](Tutorial4/Tutorial4 _后加了直方.ipynb) 中实现了使用多层感知器（MLP）做人脸检测的完整流程：数据加载、可视化、模型定义、训练循环、早停、数据增强、以及在真实图像上的滑动窗口检测与可视化。

快速开始
- 将数据文件 `faces.zip` 与测试图片（例如 `nasa-small.png` 或 `test_photo.jpg`）放在 [Tutorial4/Tutorial4 _后加了直方.ipynb](Tutorial4/Tutorial4 _后加了直方.ipynb) 所在目录下（faces.zip 不要解压）。
- 在该目录打开并运行 Notebook： [Tutorial4/Tutorial4 _后加了直方.ipynb](Tutorial4/Tutorial4 _后加了直方.ipynb)
- 主要依赖：PyTorch、scikit-image、scikit-learn、numpy、matplotlib、torchvision

主要步骤（Notebook 内）
- 加载并预处理数据（灰度、reshape、打乱）：变量 [`trainI`](Tutorial4/Tutorial4 _后加了直方.ipynb), [`testI`](Tutorial4/Tutorial4 _后加了直方.ipynb), [`trainY`](Tutorial4/Tutorial4 _后加了直方.ipynb), [`testY`](Tutorial4/Tutorial4 _后加了直方.ipynb)
- 划分训练/验证：[`vtrainI`](Tutorial4/Tutorial4 _后加了直方.ipynb), [`validI`](Tutorial4/Tutorial4 _后加了直方.ipynb), [`vtrainY`](Tutorial4/Tutorial4 _后加了直方.ipynb), [`validY`](Tutorial4/Tutorial4 _后加了直方.ipynb)
- 可视化工具：[`image_montage`](Tutorial4/Tutorial4 _后加了直方.ipynb), `show_imgs` (Notebook 中函数)
- 模型与训练：
  - 模型类：[`MyModel`](Tutorial4/Tutorial4 _后加了直方.ipynb)
  - 训练循环：[`train_model`](Tutorial4/Tutorial4 _后加了直方.ipynb)
  - 早停：[`EarlyStopping`](Tutorial4/Tutorial4 _后加了直方.ipynb)
  - 数据加载器转换器：[`get_dataloader`](Tutorial4/Tutorial4 _后加了直方.ipynb)
- 数据增强：
  - 增强流水线示例：[`augmentation`](Tutorial4/Tutorial4 _后加了直方.ipynb)
  - 增强数据集类：[`AugmentedTensorDataset`](Tutorial4/Tutorial4 _后加了直方.ipynb)
  - 增强 DataLoader：[`get_aug_dataloader`](Tutorial4/Tutorial4 _后加了直方.ipynb)
  - 增强策略集合：[`augmentations`](Tutorial4/Tutorial4 _后加了直方.ipynb)
- 最佳模型与检测：
  - 最佳训练模型：[`best_model`](Tutorial4/Tutorial4 _后加了直方.ipynb)
  - 在滑动窗口上预测并可视化检测热图（Notebook 中 `prednewY` / `prednewY2` / `prednewY3` 与绘图代码）

运行提示
- 建议先运行数据加载与展示单元，确认 `faces.zip` 能正确读取并生成 `trainI/testI`。
- 训练时设置固定随机种子以复现结果（Notebook 中已使用 torch.manual_seed 和 np.random.seed）。
- 若显存或计算慢，可减小 batch_size 或缩短 epoch，或用较小模型（hidden_sizes）。
