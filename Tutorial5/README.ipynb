# Tutorial 5 — Face Detection (简要说明)

说明  
本 Notebook 演示了使用 PyTorch 进行人脸检测的完整流程：数据加载（faces.zip）、可视化、线性模型、CNN、数据增强、训练/验证流程、以及在真实图像上的滑动窗口检测。主文件：[Tutorial5/Tutorial5.ipynb](Tutorial5/Tutorial5.ipynb)

快速开始
- 将 `faces.zip`（不要解压）和测试图片（例如 `nasa-small.png`）放到与 [Tutorial5/Tutorial5.ipynb](Tutorial5/Tutorial5.ipynb) 同一目录。
- 在该目录打开并运行 Notebook：[Tutorial5/Tutorial5.ipynb](Tutorial5/Tutorial5.ipynb)
- 建议先运行数据加载与展示单元，确认数据能正确读取并生成 `trainI` / `testI`。

主要依赖
- Python、PyTorch、torchvision、scikit-image、scikit-learn、matplotlib、numpy、scipy

主要步骤 & 关键符号（在 Notebook 中）
- 数据加载与预处理：[Tutorial5/Tutorial5.ipynb](Tutorial5/Tutorial5.ipynb)  
  - 标签与张量：[`trainI`](Tutorial5/Tutorial5.ipynb), [`testI`](Tutorial5/Tutorial5.ipynb), [`trainY`](Tutorial5/Tutorial5.ipynb), [`testY`](Tutorial5/Tutorial5.ipynb)
- 可视化工具：[`image_montage`](Tutorial5/Tutorial5.ipynb), [`show_imgs`](Tutorial5/Tutorial5.ipynb)
- 训练控制：[`EarlyStopping`](Tutorial5/Tutorial5.ipynb), [`HistoryWrapper`](Tutorial5/Tutorial5.ipynb), [`plot_history`](Tutorial5/Tutorial5.ipynb)
- 数据增强：[`add_gauss_noise`](Tutorial5/Tutorial5.ipynb), [`add_corrupt_noise`](Tutorial5/Tutorial5.ipynb), [`add_scale_shift`](Tutorial5/Tutorial5.ipynb), [`transform`](Tutorial5/Tutorial5.ipynb), [`AugmentedTensorDataset`](Tutorial5/Tutorial5.ipynb)
- 模型定义与训练：[`CNN19`](Tutorial5/Tutorial5.ipynb), [`BestCNN`](Tutorial5/Tutorial5.ipynb), [`nn_model`](Tutorial5/Tutorial5.ipynb), [`train_loader`](Tutorial5/Tutorial5.ipynb), [`valid_loader`](Tutorial5/Tutorial5.ipynb)

提示
- 使用固定随机种子以便复现（Notebook 中已有示例）。
- 训练时可先用小模型和少量 epochs 验证流程，再扩展为完整训练。
- 若显存/速度受限，可减小 batch_size 或使用更简单的网络。
